```python
# Load dataset
df = pd.read_csv('/content/TomLembong-praprepo1.csv')
```

```python
kamus = pd.read_csv('/content/full_lexicon.csv')
```

```python
print(kamus.columns)
print(df.columns)
```

```python
sentimen_dict = dict(zip(kamus['word'], kamus['weight']))
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
# Tokenisasi ulang untuk kolom 'final_cleaned_text' (jika belum tokenized)
from nltk.tokenize import word_tokenize

df['clean_text'] = df['clean_text'].fillna('').astype(str)
df['tokens'] = df['clean_text'].apply(word_tokenize)

# Sekarang kolom 'tokens' sudah bisa langsung digunakan tanpa literal_eval
print(df['tokens'].head())
```

```python
print(df.columns)
```

```python
def hitung_skor_dan_label(tokens):
    total_skor = sum(sentimen_dict.get(token.lower(), 0) for token in tokens)
    if total_skor > 0:
        label = 'positif'
    elif total_skor < 0:
        label = 'negatif'
    else:
        label = 'netral'
    return pd.Series([total_skor, label])
```

```python
df[['skor_total', 'sentimen']] = df['tokens'].apply(hitung_skor_dan_label)
```

```python
df.head()
```

```python
df.drop(['publishedAt', 'authorDisplayName', 'likeCount', 'textDisplay', 'tokens', 'skor_total'], axis=1, inplace=True)
```
